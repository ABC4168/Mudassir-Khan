Lets split the data and then independent variables are to be normalized.

```python
x = df1.iloc[:,1:]
y = df1.iloc[:,0]
```

```python
x
```

```python
y
```

```python
y = y.values.reshape(-1,1)
```

```python
y.shape
```

```python
x.skew()
```

```python
#Let's take the threshold skewwness to be 0.5
#LLog transformation is applied to remove the skewwness

for index in x.skew().index:
    if x.skew().loc[index]>0.5:
        x[index]=np.log1p(x[index])
    if x.skew().loc[index]<-0.5:
        x[index]=np.square(x[index])
```

```python
x.skew()
```

The skewwness is removed.

```python
x
```

Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
pd.DataFrame(x)
```

Standard Scaler has brought the data into single scale where it is easy to the model to understand the data.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
lr = LinearRegression()
for i in range(0,1000):
    x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=i,test_size=0.20)
    lr.fit(x_train, y_train)
    pred_train = lr.predict(x_train)
    pred_test = lr.predict(x_test)
    if round(mean_absolute_error(pred_train, y_train)*100,1) == round(mean_absolute_error(pred_test, y_test)*100,1):
        print('at random state', i, 'The model performs well')
        print('at random state', i)
        print('training socre is:', mean_absolute_error(pred_train, y_train)*100)
        print('testing score is:', mean_absolute_error(pred_test, y_test)*100)
```
